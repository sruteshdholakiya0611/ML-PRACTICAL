```python
from matplotlib import pyplot as mtp
import numpy as np
```

```python
a = np.array([1, 4, 8, 10])
b = np.array([6, 10, 22, 69])

mtp.plot(a, b, "^:m", ms = 15)
mtp.show()
```

```python
a = np.array([12, 23, 35])
b = np.array([56, 25, 45])

mtp.plot(a, b, marker = "^", ms = 10, mec = "r", mfc = "b")
mtp.show()
```

```python
a = np.array([20, 2, 64, 35])
b = np.array([54, 5, 63, 56])

mtp.plot(a, b, linestyle = "dashed", color = "g", linewidth = 10)
mtp.show()
```

```python
y = np.array([10000, 5555, 7908, 5009])
mylabels = ["Player1", "Player2", "Player3", "Player4"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "red", "yellow", "gray"]

mtp.pie(y, labels = mylabels, explode = myexplode, shadow=True, colors=mycolors, startangle=90)
mtp.legend(title="Four Players:")
mtp.show() 
```
